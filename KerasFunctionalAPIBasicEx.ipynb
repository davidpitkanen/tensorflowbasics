```python
import tensorflow as tf
import numpy as np
import random
import math

random.seed(10)
```

```python
true_m = 2.0
true_b = -4.0
```

```python
NUM_EXAMPLES = 1000 
NOISE_FACTOR = 20
xs = [random.uniform(-2 * math.pi, 2 * math.pi) for i in range(300)]


ys_pred = [math.sin(true_m * x + true_b) for x in xs]
xdata = np.array(xs)
ydata_pred = np.array(ys_pred)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# True data expectation
fig, ax = plt.subplots()
ax.scatter(xs, ydata_pred, c='tab:blue')

ax.set(xlabel='x', ylabel='y')

ax.grid()
plt.show()
```
